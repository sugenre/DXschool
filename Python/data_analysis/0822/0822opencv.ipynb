```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#파일 읽기
#open cv를 이용해서 이미지 데이터를 읽으면 numpy의 ndarray로 리턴
#흑백은 2차원, 컬러는 3차원
src1 = cv2.imread('../0821/python_opencv-main/data/lena.jpg', cv2.IMREAD_GRAYSCALE)
#print(src1)

plt.imshow(src1, cmap='gray')
plt.axis('off')
plt.show()
```

```python
src1 = cv2.imread('../0821/python_opencv-main/data/lena.jpg', cv2.IMREAD_GRAYSCALE)
src2 = np.zeros(shape=(512, 512), dtype=np.uint8) + 100
#print(src1)

#dst = src1 + src2#덧셈을 하면 오버플로우 가능성
#이미지의 색상 값은 0~255까지여야 하는데 뎃셈을 하면 256이사의 값은
#256으로 나눈 나머지 값을가지게 됨

#아래 방법으로 주로 사용함
dst = cv2.add(src1, src2) #opencv 함수를 이용해서 더하면
#256 이상의 값는 255로 설정됨

plt.imshow(src1, cmap='gray')
plt.axis('off')
plt.show()
```

```python
# 최대값과 최소값의 차이를 크게해서 화질 개선 효과를 노림
src = cv2.imread('../0821/python_opencv-main/data/minMax.jpg', cv2.IMREAD_GRAYSCALE)

(min_val, max_val, _, _) = cv2.minMaxLoc(src)
print(min_val, max_val)

ratio = 255 / (max_val - min_val)
dst = np.round((src-min_val) * ratio).astype('uint8')
(min_dst, max_dst, _, _) = cv2.minMaxLoc(dst)
print(min_dst, max_dst)

#차이를 비교
#plt.imshow(src, cmap='gray')
plt.imshow(dst, cmap='gray')
plt.axis('off')
plt.show()
```
